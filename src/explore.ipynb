```python
df = df.drop_duplicates().reset_index(drop=True)

df.info()
```

```python
columnas_object = df.select_dtypes(include=['object', 'string']).columns
columnas_object
```

```python
for col in columnas_object:
    codigos = pd.factorize(df[col])[0]
    df[col] = codigos

df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('y',axis=1)
y = df['y']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

escalador = MinMaxScaler()

x_train_esc = escalador.fit_transform(x_train,y_train)
x_test_esc = escalador.transform(x_test)

x_train_esc = pd.DataFrame(x_train_esc,index = x_train.index, columns=x_train.columns)
x_test_esc = pd.DataFrame(x_test_esc,index = x_test.index, columns=x_test.columns)
```

```python
x_test_esc
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

model = LogisticRegression()

model.fit(x_train_esc, y_train)
```

```python
y_pred = model.predict(x_test_esc)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(f"Mi precisiÃ³n es: {accuracy_score(y_test, y_pred):.4f}")
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

grid = {
        'solver': ['saga','liblinear', 'lbfgs'],
        'penalty':['elasticnet', 'l1', 'l2'],
        'C':np.logspace(-4,4,10)
}

busqueda_grilla = GridSearchCV(estimator=model,
                               param_grid = grid,
                               cv = 5,
                               scoring = 'accuracy')
```

```python
busqueda_grilla.fit(x_train_esc,y_train)
print(f'MEJORES PARAMETROS: {busqueda_grilla.best_params_}')
print(f'MEJOR ACCURACY: {busqueda_grilla.best_score_}')
```
